```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from google.colab import drive
import json

drive.mount('/content/drive')
# Read students.txt
with open('/student.json', 'r') as file:
    studentDict = json.load(file)
file.close()
# Create a list of names for the dropdown
student_names = [s['name'] for s in studentDict]

# Create the dropdown widget
dropdown = widgets.Dropdown(
    options=["Select a student"] + student_names,  # <-- Default placeholder
    value="Select a student",
    description='Student:',
    layout=widgets.Layout(width='300px')
)
# Output area
output = widgets.Output()
# Output area for submit result only
submit_output = widgets.Output()

# Helper function to find student
def findStudent(name):
  for student in studentDict:
    if student['name'] == name:
        return student
  return None

# TV show input
tv_show_input = widgets.Text(
    description='TV Show:',
    placeholder='Enter favorite show',
    layout=widgets.Layout(width='300px')
)
# Submit action
def on_submit_clicked(btn):
    submit_output.clear_output()
    selected = findStudent(dropdown.value)
    tvShow = tv_show_input.value
    tvShowData = {"TV Show": f"{tvShow}"}
    studentDict[studentDict.index(selected)].update(tvShowData)
    with open("/student.json", "w") as file:
        json.dump(studentDict,file)
    # Write to file
    file.close()
    with submit_output:
        print(f"TV Show: {tvShow} added Successfully to",selected["name"])

# When a student is selected, show their info and form
def on_student_selected(change):
    output.clear_output()
    submit_output.clear_output()
    selected = findStudent(dropdown.value)
    submit_btn = widgets.Button(
    description='Submit',
    layout=widgets.Layout(width='200px')
    )
    # Connect the button
    submit_btn.on_click(on_submit_clicked)
    with output:
        print(f"Name: ",selected["name"])
        print(f"Email: ",selected["Email"])
        print(f"Courses: ",selected["Courses"])
        print(f"Link: ",selected["Link"])
        display(tv_show_input)
    with submit_output:
        display(submit_btn)


# Attach the selection event
dropdown.observe(on_student_selected, names='value')

# Display everything
display(dropdown)
display(output)
display(submit_output)
```
