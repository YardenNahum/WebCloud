```python
import ipywidgets as widgets
from IPython.display import display, clear_output

# Read students.txt
studentsArr = []
with open('students.txt', 'r') as file:
    for line in file:
        parts = line.strip().split(',')
        studentsArr.append(parts)
file.close()
# Create a list of names for the dropdown
student_names = [s[0] for s in studentsArr]

# Create the dropdown widget
dropdown = widgets.Dropdown(
    options=["Select a student"] + student_names,  # <-- Default placeholder
    value="Select a student",
    description='Student:',
    layout=widgets.Layout(width='300px')
)
# Output area
output = widgets.Output()
# Output area for submit result only
submit_output = widgets.Output()

# Helper function to find student
def findStudent(name):
    for student in studentsArr:
        if student[0] == name:
            return student

tv_show_input = widgets.Text(
    description='TV Show:',
    placeholder='Enter favorite show',
    layout=widgets.Layout(width='300px')
)
# Submit action
def on_submit_clicked(btn):
    submit_output.clear_output()
    selected = findStudent(dropdown.value)
    tvShow = tv_show_input.value
    line = f"{selected[0]} , {selected[1]} , {selected[2]} , {selected[3]} , {tvShow}\n"

    # Write to file
    with open("students.txt", "a") as file:
        file.write(line)
    file.close()
    with submit_output:
        print("Student added successfully!")

# When a student is selected, show their info and form
def on_student_selected(change):
    output.clear_output()
    submit_output.clear_output()
    selected = findStudent(dropdown.value)
    submit_btn = widgets.Button(
    description='Submit',
    layout=widgets.Layout(width='200px')
    )
    # Connect the button
    submit_btn.on_click(on_submit_clicked)
    with output:
        print(f"Name: {selected[0]}")
        print(f"Email: {selected[1]}")
        print(f"Courses: {selected[2]}")
        print(f"Link: {selected[3]}")
        display(tv_show_input)
    with submit_output:
        display(submit_btn)


# Attach the selection event
dropdown.observe(on_student_selected, names='value')

# Display everything
display(dropdown)
display(output)
display(submit_output)
```
